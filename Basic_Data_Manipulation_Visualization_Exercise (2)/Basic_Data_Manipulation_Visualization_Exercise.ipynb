### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
#no of male per occupation/total gender by occupation

users['is_male']=np.where(users.gender=='M',1,0)

users['is_male'].sum()


male_per_occupation=pd.pivot_table(data=users,values='is_male',index='occupation',aggfunc='sum')
#while dividing ,column name should be same 
total_gender_occupation=pd.pivot_table(data=users,values='gender',index='occupation',aggfunc=len).rename(columns={'gender':'is_male'})
order_value=male_per_occupation/total_gender_occupation
order_value.sort_values(by='is_male',ascending=False)

```

```python
#sql
#select occupation,sum(is_male) from users groupBy occupation
#to make it as dataframe use []
#colum name = is_male
male_count=users.groupby('occupation')[['is_male']].agg(np.sum)
#column name = gender
total_count=users.groupby('occupation')[['gender']].agg(len).rename(columns={'gender':'is_male'})
#rename to divide
male_count/total_count
```

### 32. For each combination of occupation and gender, calculate the mean age

### 33.  For each occupation present the percentage of women and men

```python
#men
#create one column as is_male , print 1 if gender is 'M'
#male per occupation/total users per occupation
users['is_male']=np.where(users.gender =='M',1,0)

#male per occupation
male_users=users.pivot_table(values='is_male',index='occupation',aggfunc=np.sum)

#total users per occupation
total_users=users.pivot_table(values='is_male',index='occupation',aggfunc=len)

male_percent = male_users/total_users
percent_male=male_percent*100


#female
female_users =total_users-male_users
female_percent=female_users/total_users
percent_female=female_percent.rename(columns={'is_male':'is_female'})*100
percent_female


percentage_gender_users=pd.concat([percent_male,percent_female],join='inner',keys='occupation',axis=1)



percentage_gender_users


```
